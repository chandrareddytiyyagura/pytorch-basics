## Successive Differentiation

```python
from torch.autograd import grad

def nth_derivative(f, wrt, n=2):
    for i in range(n):
        grads = grad(f, wrt, create_graph=True)[0]
        f = grads.sum()

    return grads
```

```python
x= torch.tensor(5.0, requires_grad=True)
```

$f(x) = x^2 + x^3$

$f'(x) = 2x + 3x^2$

$f''(x) = 2 + 6x$

```python
f = x**2 + x**3

# calculating double derivative
nth_derivative(f, x)
```
