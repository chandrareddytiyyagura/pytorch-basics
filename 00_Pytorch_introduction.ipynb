## Other commonly used tensors

```python
x= torch.empty(size=(3,3))
x
```

```python
x = torch.zeros(size=(3,3))
x
```

```python
x = torch.ones(size=(3,3))
x
```

```python
x = torch.rand(size=(3,3))
x
```

```python
x = torch.eye(3)
x
```

```python
x =torch.arange(start=0, end=7, step=2.99)
x
```

```python
x = torch.linspace(start=0, end=7, steps=10)
x
```

```python
x = torch.rand(size=(3,4)).normal_(mean=0, std=1)
x
```

```python
x = torch.rand(size=(3,4)).uniform_(3,6)
x
```

```python
x= torch.diag(torch.ones(10))
x
```

```python
x = torch.diag(5*torch.ones(10))
x
```

```python
x = torch.arange(4)
x
```

```python
x.bool()
```

```python
x.int()
```

```python
x.short() #  int16
```

```python
x.long() # int 64
```

```python
x.float() # float32
```

```python
x.double() # float 64
```
